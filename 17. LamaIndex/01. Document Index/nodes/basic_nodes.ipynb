```python
from llama_index.node_parser import SentenceSplitter, TokenTextSplitter
from llama_index import Document

text_list = ["This is my first sample.",
             "This is my second sample.",
             "This is my third sample"]

# create document object
documents = [Document(text=t) for t in text_list]

parser = SentenceSplitter()

# create nodes from document
nodes = parser.get_nodes_from_documents(documents)

print(nodes)
```

```python
len(nodes)
```

```python
for i in nodes:
    print(i)
```

```python
from llama_index import SimpleDirectoryReader

documents = SimpleDirectoryReader("../data",).load_data()
```

```python
parser = TokenTextSplitter(chunk_size = 64, chunk_overlap = 8 )

# create nodes from document
nodes = parser.get_nodes_from_documents(documents)

print(nodes)
```
